```python
# Import Library
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
# Definisikan rentang nilai untuk input dan output
x_PE = np.arange(0, 20, 0.1)  # Jumlah Peralatan Elektronik: 0-19
x_L = np.arange(0, 26, 1)  # Jumlah Lampu: 0-25
x_D = np.arange(0, 25, 1)  # Durasi Waktu Dirumah: 0-24
#output
x_tegangan = np.arange(0, 2201, 1)  # Besar Tegangan Pemakaian: 0-2200
```

```python
# Fungsi keanggotaan Jumlah Peralatan Elektronik
mu_PE_sedikit = fuzz.trapmf(x_PE, [0, 0, 2, 10])
mu_PE_sedang = fuzz.trimf(x_PE, [2, 10, 19])
mu_PE_banyak = fuzz.trapmf(x_PE, [10, 19, 19, 19])

# Fungsi keanggotaan Jumlah Titik Lampu
mu_L_sedikit = fuzz.trapmf(x_L, [0, 0, 3, 14])
mu_L_sedang = fuzz.trimf(x_L, [3, 14, 25])
mu_L_banyak = fuzz.trapmf(x_L, [14, 25, 25, 25])

# Fungsi keanggotaan Lama Waktu Dirumah
mu_D_sebentar = fuzz.trapmf(x_D, [0, 0, 6, 15])
mu_D_sedang = fuzz.trimf(x_D, [6, 15, 24])
mu_D_lama = fuzz.trapmf(x_D, [15, 24, 24, 24])

# Fungsi keanggotaan Besar Tegangan Pemakaian
mu_tegangan_rendah = fuzz.trapmf(x_tegangan, [0, 0, 450, 1325])
mu_tegangan_sedang = fuzz.trimf(x_tegangan, [450, 1325, 2200])
mu_tegangan_tinggi = fuzz.trapmf(x_tegangan, [1325, 2200, 2200, 2200])
```

```python
# Plot fungsi keanggotaan untuk input dan output
plt.figure(figsize=(10, 6))
plt.plot(x_PE, mu_PE_sedikit, label='Sedikit (PE)')
plt.plot(x_PE, mu_PE_sedang, label='Sedang (PE)')
plt.plot(x_PE, mu_PE_banyak, label='Banyak (PE)')
plt.title('Fungsi Keanggotaan Jumlah Peralatan Elektronik')
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x_L, mu_L_sedikit, label='Sedikit (L)')
plt.plot(x_L, mu_L_sedang, label='Sedang (L)')
plt.plot(x_L, mu_L_banyak, label='Banyak (L)')
plt.title('Fungsi Keanggotaan Jumlah Titik Lampu')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x_D, mu_D_sebentar, label='Sebentar (D)')
plt.plot(x_D, mu_D_sedang, label='Sedang (D)')
plt.plot(x_D, mu_D_lama, label='Lama (D)')
plt.title('Fungsi Keanggotaan Durasi Waktu Dirumah')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x_tegangan, mu_tegangan_rendah, label='Rendah (Tegangan)')
plt.plot(x_tegangan, mu_tegangan_sedang, label='Sedang (Tegangan)')
plt.plot(x_tegangan, mu_tegangan_tinggi, label='Tinggi (Tegangan)')
plt.title('Fungsi Keanggotaan Besar Tegangan Pemakaian')
plt.legend()
plt.grid()
plt.show()
```

## Fungsi Keanggotaan Jumlah Peralatan Listrik
$\mu_{[PE\_{\text{sedikit}}]} = \begin{cases} 
1 & \text{jika } PE \leq 2 \\
\frac{10 - PE}{10 - 2} & \text{jika } 2 < PE < 10 \\
0 & \text{jika } PE \geq 10 
\end{cases}$

$\mu_{[PE\_{\text{sedang}}]} = \begin{cases}
0 & \text{jika } PE \leq 2 \text{ atau } PE \geq 10 \\
\frac{PE - 2}{10 - 2} & \text{jika } 2 < PE < 10 \\
\frac{19 - PE}{19 - 10} & \text{jika } 10 < PE < 19 
\end{cases}$

$\mu_{[PE\_{\text{banyak}}]} = \begin{cases}
0 & \text{jika } PE \leq 10 \\
\frac{PE - 10}{19 - 10} & \text{jika } 10 < PE < 19 \\
1 & \text{jika } PE \geq 19 
\end{cases}$

## Fungsi Keanggotaan Jumlah Lampu
$\mu_{[L\_{\text{sedikit}}]} = \begin{cases}
1 & \text{jika } L \leq 3 \\
\frac{14 - L}{14 - 3} & \text{jika } 3 < L < 10 \\
0 & \text{jika } L \geq 14 
\end{cases}$

$\mu_{[L\_{\text{sedang}}]} = \begin{cases}
0 & \text{jika } L \leq 3 \text{ atau } L \geq 25 \\
\frac{L - 3}{14 - 3} & \text{jika } 3 < L < 14 \\
\frac{25 - L}{25 - 14} & \text{jika } 14 < L < 25 
\end{cases}$

$\mu_{[L\_{\text{banyak}}]} = \begin{cases}
0 & \text{jika } L \leq 14 \\
\frac{L - 14}{25 - 14} & \text{jika } 14 < L < 25 \\
1 & \text{jika } L \geq 25 
\end{cases}$

## Fungsi Keanggotaan Lama Durasi Pemakaian
$\mu_{[D\_{\text{sebentar}}]} = \begin{cases}
1 & \text{jika } D \leq 6 \\
\frac{15 - D}{15 - 6} & \text{jika } 6 < D < 15 \\
0 & \text{jika } D \geq 15 
\end{cases}$

$\mu_{[D\_{\text{sedang}}]} = \begin{cases} 
0 & \text{jika } D \leq 6 \text{ atau } D \geq 24 \\
\frac{D - 6}{15 - 6} & \text{jika } 6 < D < 15 \\
\frac{24 - D}{24 - 15} & \text{jika } 15 < D < 24 
\end{cases}$

$\mu_{[D\_{\text{lama}}]} = \begin{cases}
0 & \text{jika } D \leq 15 \\
\frac{D - 15}{24 - 15} & \text{jika } 15 < D < 24 \\
1 & \text{jika } D \geq 24 
\end{cases}$


```python
# Fungsi keanggotaan Jumlah Peralatan Elektronik
def PE_sedikit(pe):
    if pe <= 2:
        return 1
    elif 2 < pe < 10:
        return (10 - pe) / (10 - 2)
    else:
        return 0

def PE_sedang(pe):
    if pe <= 2 or pe >= 10:
        return 0
    elif 2 < pe < 10:
        return (pe - 2) / (10 - 2)
    elif 10 < pe < 19:
        return (19 - pe) / (19 - 10)

def PE_banyak(pe):
    if pe <= 10:
        return 0
    elif 10 < pe < 19:
        return (pe - 10) / (19 - 10)
    else:
        return 1

# Fungsi keanggotaan Jumlah Titik Lampu
def L_sedikit(L):
    if L <= 3:
        return 1
    elif 3 < L < 14:
        return (14 - L) / (14 - 3)
    else:
        return 0

def L_sedang(L):
    if L <= 3 or L >= 25:
        return 0
    elif 3 < L < 14:
        return (L - 3) / (14 - 3)
    elif 14 < L < 25:
        return (25 - L) / (25 - 14)

def L_banyak(L):
    if L <= 14:
        return 0
    elif 14 < L < 25:
        return (L - 14) / (25 - 14)
    else:
        return 1

# Fungsi keanggotaan Durasi Waktu Dirumah
def D_sebentar(D):
    if D <= 6:
        return 1
    elif 6 < D < 15:
        return (15 - D) / (15 - 6)
    else:
        return 0

def D_sedang(D):
    if D <= 6 or D >= 24:
        return 0
    elif 6 < D < 15:
        return (D - 6) / (15 - 6)
    elif 15 < D < 24:
        return (24 - D) / (24 - 15)

def D_lama(D):
    if D <= 15:
        return 0
    elif 15 < D < 24:
        return (D - 15) / (24 - 15)
    else:
        return 1
```

## Masukkan nilai yang akan diukur untuk setiap parameter

```python
# Data Uji: nilai keanggotaan untuk setiap input
data_uji = {
    'PE': int(input('Masukkan jumlah Peralatan Elektronik: ')),
    'L': int(input('Masukkan jumlah Lampu: ')), 
    'D': int(input('Masukkan lama Durasi Pemakaian: ')) #jam
}
```

```python
# Menghitung nilai keanggotaan untuk inputan
mu_PE = {
    'sedikit': PE_sedikit(data_uji['PE']),
    'sedang': PE_sedang(data_uji['PE']),
    'banyak': PE_banyak(data_uji['PE']),
}

mu_L = {
    'sedikit': L_sedikit(data_uji['L']),
    'sedang': L_sedang(data_uji['L']),
    'banyak': L_banyak(data_uji['L']),
}

mu_D = {
    'sebentar': D_sebentar(data_uji['D']),
    'sedang': D_sedang(data_uji['D']),
    'lama': D_lama(data_uji['D']),
}

# Menampilkan nilai keanggotaan
print(f"Keanggotaan Jumlah Peralatan Elektronik: {mu_PE}")
print(f"Keanggotaan Jumlah Lampu: {mu_L}")
print(f"Keanggotaan Durasi Waktu Dirumah: {mu_D}")

# Menentukan nilai output untuk setiap aturan (rendah, sedang, tinggi)
output_values = {
    'rendah': 450,
    'sedang': 1325,
    'tinggi': 2200
}
```

```python
# Aturan fuzzy
def tsukamoto_inference(data_uji):
    rules = [
        ('sedikit', 'sedikit', 'sebentar', 'rendah'),
        ('sedikit', 'sedikit', 'sedang', 'rendah'),
        ('sedikit', 'sedikit', 'lama', 'rendah'),
        ('sedikit', 'sedang', 'sebentar', 'rendah'),
        ('sedikit', 'sedang', 'sedang', 'sedang'),
        ('sedikit', 'sedang', 'lama', 'sedang'),
        ('sedikit', 'banyak', 'sebentar', 'rendah'),
        ('sedikit', 'banyak', 'sedang', 'sedang'),
        ('sedikit', 'banyak', 'lama', 'sedang'),
        ('sedang', 'sedikit', 'sebentar', 'rendah'),
        ('sedang', 'sedikit', 'sedang', 'sedang'),
        ('sedang', 'sedikit', 'lama', 'sedang'),
        ('sedang', 'sedang', 'sebentar', 'sedang'),
        ('sedang', 'sedang', 'sedang', 'sedang'),
        ('sedang', 'sedang', 'lama', 'sedang'),
        ('sedang', 'banyak', 'sebentar', 'sedang'),
        ('sedang', 'banyak', 'sedang', 'sedang'),
        ('sedang', 'banyak', 'lama', 'tinggi'),
        ('banyak', 'sedikit', 'sebentar', 'sedang'),
        ('banyak', 'sedikit', 'sedang', 'sedang'),
        ('banyak', 'sedikit', 'lama', 'tinggi'),
        ('banyak', 'sedang', 'sebentar', 'sedang'),
        ('banyak', 'sedang', 'sedang', 'tinggi'),
        ('banyak', 'sedang', 'lama', 'tinggi'),
        ('banyak', 'banyak', 'sebentar', 'tinggi'),
        ('banyak', 'banyak', 'sedang', 'tinggi'),
        ('banyak', 'banyak', 'lama', 'tinggi')
    ]

    
    total_numerator = 0
    total_denominator = 0

    for rule in rules:
        # Ambil nilai keanggotaan untuk setiap bagian dari aturan
        mu_PE_val = mu_PE[rule[0]]
        mu_L_val = mu_L[rule[1]]
        mu_D_val = mu_D[rule[2]]

        # Derajat keanggotaan untuk aturan ini
        rule_strength = min(mu_PE_val, mu_L_val, mu_D_val)

        # Nilai output berdasarkan aturan
        output_value = output_values[rule[3]]

        # Akumulasi pembobotan
        total_numerator += rule_strength * output_value
        total_denominator += rule_strength

    # Menghitung nilai output (tegangan) menggunakan rata-rata tertimbang
    z = total_numerator / total_denominator if total_denominator != 0 else 0
    return z

```

```python
# Menampilkan hasil output
output_tegangan = tsukamoto_inference(data_uji)
print(f"Nilai Tegangan Pemakaian Listrik: {output_tegangan:.2f} V")
```
